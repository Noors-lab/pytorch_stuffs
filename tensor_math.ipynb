# the libraries


```python
import torch
```

```python
tensor_a = torch.tensor([1,2,3,4])
tensor_b = torch.tensor([5,6,7,8])
```

```python
# addition

tensor_a + tensor_b
```

```python
# addtion longhand

torch.add(tensor_a,tensor_b)
```

```python
# substraction
tensor_b - tensor_a
```

```python
# sub function
torch.sub(tensor_b , tensor_a)
```

```python
# multiplication

tensor_a * tensor_b
```

```python
# mul longhand
torch.mul(tensor_a , tensor_b)
```

```python
# division
tensor_b / tensor_a
```

```python
# div longhand
torch.div(tensor_b,tensor_a)
```

```python
# remainder module
tensor_b % tensor_a
```

```python
# remainder longhand
torch.remainder(tensor_b,tensor_a)
```

```python
# exponents
tensor_a ** tensor_b
```

```python
# exponents longhand
torch.pow(tensor_a,tensor_b)
```

```python
# another way of longhand
tensor_a.add(tensor_b)
```

```python
# another way of longhand
tensor_b.sub(tensor_a)
```
